```python
import astropy.units as u
import astropy.constants as ac
import numpy as np
```

How to define a function?

```python
  def calculatingOrbitalVelocity(M, R):
   """
   This is a function that calculates the orbital velocity
   Input:
   M, the mass of the central objects
   R, the orbital radius of satellite
   Output:
   orbitbalVelovity, the orbital velocity of the satellite
   """
   orbitalVelovity = np.sqrt(ac.G * M / R)
   return orbitalVelovity
```

```python
# Below we define input M and R
M = 1 * u.earthMass
R = 1 * u.earthRad
# Calculate orbital velocity of an earth satellite
orbitbalVelovity = calculatingOrbitalVelocity(M, R)
print(orbitbalVelovity.to(u.km/u.s)) # using the predefined function
```

Introducing the concept of an array

```python
# Orbital distances of the eight solar system planets from inside out
#Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune
R_planet_arr = np.array([0.4,0.7,1.0,1.5,5.2,9.6,19.2,30.1]) * u.AU
```

```python
# Now we calculate the orbital velocities for eight planets
M_sun = 1 * u.solMass
v_planet_arr = calculatingOrbitalVelocity(M_sun, R_planet_arr)
print(v_planet_arr.to(u.km/u.s))
```
