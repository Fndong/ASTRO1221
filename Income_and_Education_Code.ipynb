```python
#import neccisary functions
import numpy as np
import matplotlib.pyplot as plt
```

```python
F_inc_arr = np.array([30900,36700,39100,43100,60800,75700]) # Create an array for the median annual full time salary
Education_arr = np.arange(len(F_inc_arr)) # Create an array for education levels
plt.plot (Education_arr, F_inc_arr) # Plot Education by Income
plt.xlabel("Education")
plt.ylabel("Female Income")
plt.show()
```

```python
z_linear = np.polyfit(Education_arr, F_inc_arr, 1) #Create linear euqation using polyfit function
print("Slope =", z_linear[0], "in dollar per level") #Print slope of linear euqation
print("Y-intercept = ", z_linear[1],"in dollar at level 0 since...") #Print y-int of linear equation
```

```python
z_quadratic = np.polyfit(Education_arr, F_inc_arr, 2) # Create a quadratic formula using polyfit function
print(z_quadratic)
print("That is: Income - {:.1f}* level**2 + {:.1f} * level + {:.1f}".format(z_quadratic[0],z_quadratic[1],z_quadratic[2]))
```

```python
O = F_inc_arr # set O to the female income array values
O_err = np.array([3310.,960.,3070.,2580.,860.,4940., ])# create an array of the confidence interval values
C_linear = z_linear[0] * Education_arr + z_linear[1] # defined values to input into the linear equation
C_quadratic = z_quadratic[0] * Education_arr**2 + z_quadratic[1] * Education_arr + z_quadratic[2] # defined values to put into the quadratic equation
```

```python
#plot the linear and quadratic equations with the plotted income values
plt.errorbar(Education_arr, O, yerr=O_err, marker="o", linestyle = "None", color = "black", label = "Data")
plt.plot(Education_arr,C_linear, color = "red", label = "Linear fit")
plt.plot(Education_arr,C_quadratic, color = "blue", label = "Quadratic fit")
plt.legend()
plt.xlabel("Education Level")
plt.ylabel("Female Income [$]")
plt.show()
```
